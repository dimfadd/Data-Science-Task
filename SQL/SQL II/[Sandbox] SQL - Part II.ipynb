```python
%%sql
WITH invoices_date_breakdown_2013 AS (
  SELECT
    InvoiceId,
    InvoiceDate,
    DATE_PART('year', InvoiceDate) AS year,
    DATE_PART('month', InvoiceDate) AS month,
    BillingCity,
    Total
  FROM
    invoices
  WHERE
    DATE_PART('year', InvoiceDate) = 2013
),

agg_per_month_2013 AS (
  SELECT
    BillingCity,
    month,
    SUM(Total) AS total_per_month_2013
  FROM
    invoices_date_breakdown_2013
  GROUP BY
    BillingCity,
    month
  ORDER BY
    BillingCity ASC,
    month ASC
)

SELECT * FROM agg_per_month_2013
```

```python
%%sql
SELECT
  c.CustomerId,
  CONCAT(c.FirstName, ' ', c.LastName) AS cust_name,
  e.EmployeeId,
  CONCAT(e.FirstName, ' ', e.LastName) AS employee_name
FROM
  customers c
LEFT JOIN
  employees e
ON
  c.SupportRepId = e.EmployeeId;
```

```python
%%sql
WITH cust_employee_pair AS (
  SELECT
    c.CustomerId,
    CONCAT(c.FirstName, ' ', c.LastName) AS cust_name,
    e.EmployeeId,
    CONCAT(e.FirstName, ' ', e.LastName) AS employee_name
  FROM
    customers c
  LEFT JOIN
    employees e
  ON
    c.SupportRepId = e.EmployeeId
)

SELECT
  EmployeeId,
  employee_name,
  COUNT(CustomerId) AS cust_cnt
FROM
  cust_employee_pair
GROUP BY
  EmployeeId, employee_name
```
