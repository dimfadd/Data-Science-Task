```python
from pandas.core.interchange.dataframe_protocol import Column
paid_transaction = pd.read_excel('/content/drive/MyDrive/Digital skola/dataframe II/Homework - DataFrame.xlsx' ,sheet_name='Paid-Transaction')
paid_transaction.columns = paid_transaction.iloc[0]
Paid_Transaction = paid_transaction.drop([0])
Paid_Transaction = Paid_Transaction.loc[:, Paid_Transaction.columns.notna()]
Paid_Transaction = Paid_Transaction.reset_index(drop=True)
Paid_Transaction.info()
```

```python
from pandas.core.interchange.dataframe_protocol import Column
Transaction_Fee = pd.read_excel('/content/drive/MyDrive/Digital skola/dataframe II/Homework - DataFrame.xlsx' ,sheet_name='Transaction Fee')
Transaction_Fee.columns = Transaction_Fee.iloc[3]
Transaction_Fee = Transaction_Fee.drop([0,1,2,3])
Transaction_Fee = Transaction_Fee.loc[:, Transaction_Fee.columns.notna()]
Transaction_Fee = Transaction_Fee.reset_index(drop=True)
Transaction_Fee.info()
```

```python
Paid_Transaction['Paid Date'] = Paid_Transaction['Paid Date'].astype(str).str[:-2]
Paid_Transaction
```

```python
grupby = Paid_Transaction.groupby([pd.Grouper(key='Paid Date'), pd.Grouper(key='Meta Category')])[['Transaction Amount', 'Sales Discount','Other Discount']].sum().reset_index()
#grupby = Paid_Transaction.groupby(pd.Grouper(key='Paid Date'))[['Transaction Amount', 'Sales Discount','Other Discount']].sum().reset_index()
grupby
```

```python
# Merge the two dataframes on the "Month" column
#mer = Transaction_Fee.merge(grupby, left_index=True, right_on='Paid Date', how='right')
merged_df = pd.concat([grupby, Transaction_Fee], axis=1)
merged_df
```

```python
# Calculate the transaction fee amount
merged_df["Transaction Fee Amount"] = merged_df["Transaction Fee Rate"] * merged_df["Transaction Amount"]
# Calculate the net revenue
merged_df["Net Revenue"] = merged_df["Transaction Fee Amount"] - merged_df["Sales Discount"] - merged_df["Other Discount"]
# Print result
merged_df
```
