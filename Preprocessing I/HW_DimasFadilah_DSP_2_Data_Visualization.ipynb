## Columns Renaming

```python
rename_mapper = {col:col.replace(" ", "_").lower() for col in df.columns}
df.rename(
    columns=rename_mapper,
    inplace=True
)

df.head()
```

## To Datetime datatype

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head().info()
```

```python
df = df.sort_values(
    by=["order_id", "order_date", "ship_date"],
    ascending=[True, True, True],
    ignore_index=True
)

df.head(20)
```

## Remove "$" in ["sales", "profit", "shipping_cost" ]

```python
dollar_cols = ["sales", "profit", "shipping_cost"]
for col in dollar_cols:
  df[col] = (
      df[col]
      .str.replace("[^\d.]","", regex =True)
  )
  df.rename(columns={col: col+"($)"}, inplace=True)
print("Formatting & Renaming, Done!")
```

## Monthly Sales Trends

```python
monthly_sum_sales = df.groupby("months").agg({"sales($)": "sum"})
monthly_sum_sales
```

```python
ordered_months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

Month_CI = pd.CategoricalIndex(
    monthly_sum_sales.index,
    ordered=True,
    categories=ordered_months
)

Month_CI
```

```python
monthly_sum_sales.index = Month_CI
monthly_sum_sales = monthly_sum_sales.sort_index()
monthly_sum_sales
```

```python
import matplotlib.ticker as ticker
```

```python
# Create line chart for monthly sales
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    monthly_sum_sales.index,
    monthly_sum_sales["sales($)"],
    marker='*',
    mfc='red', # marker face color
    mec='red', # marker edge color
    linestyle='-'
)

ax.set_title('Monthly Sales Trends')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))

# Show the line chart
fig.show()
```

## 1.&nbsp; **Monthly Trends: Plot line charts or bar charts to visualize how profit, or quantity changes month by month.**



```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

df = pd.read_csv("ecommerce_preprocessed.csv")
df
```

```python
df.columns
```

```python
monthly_profit = df.groupby('months')['profit($)'].sum()
monthly_quantity = df.groupby('months')['quantity'].sum()
```

```python
# Line Charts
plt.figure(figsize=(12, 6))
plt.plot(monthly_profit.index, monthly_profit.values, label='Total Profit($)', color='#488f31', marker='o', linestyle='-')
plt.plot(monthly_quantity.index, monthly_quantity.values, label='Total Quantity', color='#ff6361', marker='x', linestyle='--')
plt.xlabel('Bulan')
plt.ylabel('Jumlah')
plt.title('Perubahan Laba dan Kuantitas dari Bulan ke Bulan')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Mendapatkan daftar Profit dari Bulan ke Bulan
months = monthly_profit.index.astype(str)

# Membuat bar charts
plt.figure(figsize=(12, 6))
plt.bar(months, monthly_profit, color='#488f31', alpha=1)
plt.xlabel('Bulan')
plt.ylabel('Total Profit')
plt.title('Total Profit dari Bulan ke Bulan')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Membuat bar charts Quantity dari Bulan ke Bulan
plt.figure(figsize=(12, 6))
plt.bar(months, monthly_quantity, color='#ff6361', alpha=1)
plt.xlabel('Bulan')
plt.ylabel('Total Quantity')
plt.title('Total Quantity dari Bulan ke Bulan')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 2.&nbsp; **Region-wise Analysis: Create bar charts or stacked bar charts to compare sales, profit, or quantity across different regions.**

```python
region_sales = df.groupby('region')['sales($)'].sum()
region_profit = df.groupby('region')['profit($)'].sum()
region_quantity = df.groupby('region')['quantity'].sum()
```

```python
# Stacked bar charts
plt.figure(figsize=(12, 6))

x = range(len(region_sales))
width = 0.2  # Lebar setiap batang

plt.bar(x, region_sales.values, width=width, label='Penjualan($)', color='skyblue')
plt.bar(x, region_profit.values, width=width, label='Laba($)', color='orange', bottom=region_sales.values)
plt.bar(x, region_quantity.values, width=width, label='Kuantitas', color='green', bottom=region_sales.values + region_profit.values)

plt.xlabel('Region')
plt.ylabel('Jumlah')
plt.title('Perbandingan Penjualan, Laba, dan Kuantitas Berdasarkan Region')
plt.legend()
plt.xticks(x, region_sales.index)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 3.&nbsp; **Sales vs. Quantity/Discount: Create scatter plots to visualize the relationship between sales and quantity or discount. You can use color-coding or size of points to represent different product categories.**

```python
# Encoding kolom 'product_category' menjadi angka
category_mapping = {category: i for i, category in enumerate(df['product_category'].unique())}
df['category_code'] = df['product_category'].map(category_mapping)

# Definisikan variabel-variabel untuk scatter plot
x = df['sales($)']
y = df['quantity']
colors = df['category_code']

# Buat scatter plot dengan color-coding
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, cmap='viridis', alpha=0.6, edgecolor='k', s=100)

# Sesuaikan plot
plt.xlabel('Sales($)')
plt.ylabel('Quantity')
plt.title('Scatter Plot: Sales vs. Quantity (Color-coded by Product Category)')
plt.xticks(rotation=45)

# Buat label pada sumbu warna (colorbar) dengan nama kategori produk
colorbar = plt.colorbar(scatter)
colorbar.set_label('Kategori Produk')
colorbar.set_ticks(np.arange(len(df['product_category'].unique())))
colorbar.set_ticklabels(df['product_category'].unique())

plt.grid(True)

# Tampilkan plot
plt.show()

```

## 4.&nbsp; **Product Category vs. Sales/Profit: Use a bar chart or a grouped bar chart to compare sales or profit across different product categories**

```python
category_sales = df.groupby('product_category')['sales($)'].sum().sort_values(ascending=False)
category_profit = df.groupby('product_category')['profit($)'].sum().sort_values(ascending=False)
```

```python
# Bar chart
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar', color='#2f4b7c', label='Penjualan($)')
category_profit.plot(kind='bar', color='#488f31', label='Laba($)')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah($)')
plt.title('Perbandingan Penjualan dan Laba Berdasarkan Kategori Produk')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))
plt.show()

```

```python
# Grouped bar chart
plt.figure(figsize=(12, 6))
width = 0.4  # Lebar setiap batang

x = range(len(category_sales))
plt.bar(x, category_sales.values, width=width, color='#2f4b7c', label='Sales($)')
plt.bar([i + width for i in x], category_profit.values, width=width, color='#488f31', label='Profit($)')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah($)')
plt.title('Perbandingan Sales dan Profit Berdasarkan Kategori Produk')
plt.legend()
plt.xticks([i + width/2 for i in x], category_sales.index, rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))
plt.show()
```
