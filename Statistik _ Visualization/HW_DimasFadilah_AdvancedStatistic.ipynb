# 1. Tentukan H0 & H1 untuk studi kasus di atas

Dalam studi kasus ini, kita ingin menguji apakah panjang penggaris pada kuartal kedua masih memenuhi kriteria "lulus Quality Checking," yang berarti panjang penggaris masih di sekitar 30 cm. Untuk melakukan uji hipotesis, kita perlu merumuskan hipotesis null (H0) dan hipotesis alternatif (H1) sebagai berikut:

* Hipotesis Null (H0):

H0: Rata-rata panjang penggaris pada kuartal kedua sama dengan 30 cm (μ = 30). \

* Hipotesis Alternatif (H1):

H1: Rata-rata panjang penggaris pada kuartal kedua tidak sama dengan 30 cm (μ ≠ 30).


#2. Tarik kesimpulan terhadap studi kasus di atas dengan menggunakan uji hipotesis. Gunakan nilai alpha 0.05. (Gunakan library statistik scipy)

```python
import numpy as np
from scipy import stats

# Nilai alpha (tingkat signifikansi)
alpha = 0.05

# Rata-rata sampel
sample_mean = np.mean(ruler_cm)

# Standar deviasi sampel
sample_std = np.std(ruler_cm, ddof=1)

# Jumlah sampel
n = len(ruler_cm)

# Rata-rata populasi yang diasumsikan dalam H0
population_mean = 30.0

# Menghitung t-statistik
t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Menghitung derajat kebebasan
degrees_of_freedom = n - 1

# Menghitung p-value dua sisi
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), degrees_of_freedom))

# Menentukan apakah H0 dapat ditolak atau tidak
if p_value < alpha:
    print("Hipotesis Null (H0) dapat ditolak.")
else:
    print("Hipotesis Null (H0) tidak dapat ditolak.")

# Menampilkan t-statistik dan p-value
print("T-Statistik:", t_statistic)
print("P-Value:", p_value)
```


p-value lebih besar dari alpha, kita tidak dapat menolak Hipotesis Null (H0)
dan menyimpulkan bahwa tidak ada cukup bukti untuk menyatakan bahwa
rata-rata panjang penggaris berbeda dari 30 cm.
